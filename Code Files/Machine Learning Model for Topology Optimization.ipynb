## Traditional Approach

The traditional topology optimization approaches have been widely utilized in structural design optimization, employing mathematical optimization techniques, finite element analysis (FEA), and density-based methods. Among these methods, several notable techniques have emerged, including the Evolutionary Structural Optimization (ESO), Bi-directional Evolutionary Structural Optimization (BESO), Solid Isotropic Material with Penalization (SIMP), and Level Set methods.

**Evolutionary Structural Optimization (ESO)** is a pioneering method in topology optimization, aiming to discover the optimal material distribution within a given design domain. The ESO algorithm iteratively eliminates inefficient material elements based on stress analysis, leading to more efficient designs. However, ESO may face computational challenges, particularly in large-scale optimization problems, limiting its application to complex design scenarios.

**Bi-directional Evolutionary Structural Optimization (BESO)** is an extension of ESO, introducing the ability to add material in addition to removal during the optimization process. BESO offers increased flexibility in handling design constraints and generating intricate structural configurations. Nevertheless, similar to ESO, BESO's computational cost may become prohibitive for large-scale problems and intricate design spaces.

**Solid Isotropic Material with Penalization (SIMP)** is another widely used traditional method in topology optimization. The SIMP approach penalizes intermediate densities to encourage designs with fewer materials, resulting in more efficient and lightweight structures. However, SIMP may struggle to produce designs with complex geometries, and its reliance on density-based modeling can lead to numerical instabilities in certain cases.

**Level Set methods** have emerged as a versatile framework in topology optimization. By representing the interface between solid and void regions as a level set function, these methods can efficiently capture and evolve geometric shapes during the optimization process. Level Set methods provide the advantage of handling topological changes in the design efficiently. However, they may still face challenges in handling complex design spaces and require careful tuning of parameters to avoid undesired numerical artifacts.

**Limitations of the Traditional Approach**

While traditional topology optimization approaches have made significant contributions to structural design optimization, they do have certain limitations. One major limitation is their computational complexity, which can be a barrier to optimizing large-scale and complex design problems. The reliance on density-based methods in ESO and SIMP may lead to checkerboard patterns in the optimized designs, affecting the structural integrity and manufacturability of the final product.

Furthermore, traditional approaches may struggle to handle designs with intricate geometries, such as freeform structures or those involving multiple materials. Their formulation often lacks the ability to consider manufacturability constraints or other real-world design constraints that are critical in practical engineering applications.

Despite their effectiveness in some scenarios, traditional methods may not be well-suited to handle complex design spaces and intricate structural geometries. As a result, they may not fully exploit the potential for achieving the most efficient and innovative design solutions.

Overall, the limitations of traditional topology optimization methods motivate the exploration of alternative approaches, such as machine learning (ML) and deep learning (DL) techniques, which offer the prospect of revolutionizing the field by integrating historical data and advanced algorithms to model the intricate relationships between design parameters and structural responses.

## First Neural Network with UNet which uses entire dataset for training

The below code defines a neural network (UNet) and uses the entire dataset for training the model. The training function train_model iterates over the entire dataset in each epoch to update the model's parameters based on the computed loss between the predicted outputs and the ground truth targets. The training process continues for the specified number of epochs, and the learned weights of the model are saved after training is complete.

````python
# import libraries
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import h5py
import numpy as np
import matplotlib.pyplot as plt

# Define the UNet model
class UNet(nn.Module):
    def __init__(self):
        super(UNet, self).__init__()
        # Define a convolutional block
        def conv_block(in_channels, out_channels):
            return nn.Sequential(
                nn.Conv2d(in_channels, out_channels, 3, padding=1),
                nn.ReLU(),
                nn.Conv2d(out_channels, out_channels, 3, padding=1),
                nn.ReLU()
            )

        # Define the encoder and decoder layers
        self.encoder1 = conv_block(100, 16)
        self.pool1 = nn.MaxPool2d(2)
        self.encoder2 = conv_block(16, 32)
        self.pool2 = nn.MaxPool2d(2)
        self.middle = conv_block(32, 64)
        self.up2 = nn.ConvTranspose2d(64, 32, kernel_size=2, stride=2)
        self.decoder2 = conv_block(64, 32)
        self.up1 = nn.ConvTranspose2d(32, 16, kernel_size=2, stride=2)
        self.decoder1 = conv_block(32, 16)
        self.output = nn.Conv2d(16, 1, 1)

    def forward(self, x):
        # Encoder layers
        enc1 = self.encoder1(x)
        enc2 = self.encoder2(self.pool1(enc1))
        middle = self.middle(self.pool2(enc2))
        
        # Decoder layers with skip connections
        dec2 = self.decoder2(torch.cat([self.up2(middle), enc2], dim=1))
        dec1 = self.decoder1(torch.cat([self.up1(dec2), enc1], dim=1))
        
        # Output layer
        return self.output(dec1)



# Define the dataset class
class TopologyOptimizationDataset(Dataset):
    def __init__(self, dataset_path):
        with h5py.File(dataset_path, 'r') as h5f:
            self.iters = np.transpose(h5f['iters'], (0, 3, 1, 2))  # Move the iterations dimension
            self.targets = np.transpose(h5f['targets'], (0, 3, 1, 2))  # Move the channels dimension

    def __len__(self):
        return len(self.iters)

    def __getitem__(self, index):
        iter = self.iters[index]
        target = self.targets[index]
        return torch.from_numpy(iter).float(), torch.from_numpy(target).float()

# Define the training function
def train_model(model, dataset, batch_size=64, num_epochs=20, learning_rate=0.001):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    data_loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

    print('Training...')
    for epoch in range(num_epochs):
        for inputs, targets in data_loader:
            inputs = inputs.to(device)
            targets = targets.to(device)

            # Forward pass
            outputs = model(inputs)

            # Compute the loss
            loss = criterion(outputs, targets)

            # Backward and optimize
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item()}')

    print('Training complete.')

# Load the dataset
dataset = TopologyOptimizationDataset('top_dataset.h5')

# Create the model
model = UNet()

# Train the model
train_model(model, dataset)

# Save the learned weights
torch.save(model.state_dict(), 'model_weights.pth')

# Example usage for prediction
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
test_input, target = dataset[0]
test_input = test_input.to(device)  # Add this line
model.eval()
with torch.no_grad():
    prediction = model(test_input.unsqueeze(0))
    binary_prediction = (prediction > 0.5).float()  # Apply threshold to convert to binary

print('Prediction shape:', prediction.shape)
print('Binary prediction shape:', binary_prediction.shape)

# Visualize the test input, prediction, and ground truth
test_input = test_input[59, :, :].squeeze().cpu().numpy()  # Select the 60th iteration
binary_prediction = binary_prediction.squeeze().cpu().numpy()
ground_truth = target.squeeze().cpu().numpy()

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(test_input, cmap='gray')
ax[0].set_title('Test Input')
ax[1].imshow(1-binary_prediction, cmap='binary')
ax[1].set_title('Binary Prediction')
ax[2].imshow(ground_truth, cmap='gray')
ax[2].set_title('Ground Truth')
plt.tight_layout()
plt.show()
````

### Second Neural Network: with UNet which splits the dataset into test, train and validation set. It will print the accuracies and the output prediction also.

````python
import random
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
import h5py
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, roc_auc_score

# Define the UNet model
class UNet(nn.Module):
    def __init__(self):
        super(UNet, self).__init__()
        def conv_block(in_channels, out_channels):
            return nn.Sequential(
                nn.Conv2d(in_channels, out_channels, 3, padding=1),
                nn.ReLU(),
                nn.Conv2d(out_channels, out_channels, 3, padding=1),
                nn.ReLU()
            )

        self.encoder1 = conv_block(100, 16)
        self.pool1 = nn.MaxPool2d(2)
        self.encoder2 = conv_block(16, 32)
        self.pool2 = nn.MaxPool2d(2)
        self.middle = conv_block(32, 64)
        self.up2 = nn.ConvTranspose2d(64, 32, kernel_size=2, stride=2)
        self.decoder2 = conv_block(64, 32)
        self.up1 = nn.ConvTranspose2d(32, 16, kernel_size=2, stride=2)
        self.decoder1 = conv_block(32, 16)
        self.output = nn.Conv2d(16, 1, 1)

    def forward(self, x):
        enc1 = self.encoder1(x)
        enc2 = self.encoder2(self.pool1(enc1))
        middle = self.middle(self.pool2(enc2))
        dec2 = self.decoder2(torch.cat([self.up2(middle), enc2], dim=1))
        dec1 = self.decoder1(torch.cat([self.up1(dec2), enc1], dim=1))
        return self.output(dec1)


class TopologyOptimizationDataset(Dataset):
    def __init__(self, dataset_path):
        with h5py.File(dataset_path, 'r') as h5f:
            self.iters = np.transpose(h5f['iters'], (0, 3, 1, 2))
            self.targets = np.transpose(h5f['targets'], (0, 3, 1, 2))

    def __len__(self):
        return len(self.iters)

    def __getitem__(self, index):
        iter = self.iters[index]
        target = self.targets[index]
        return torch.from_numpy(iter).float(), torch.from_numpy(target).float()


def train_model(model, train_loader, val_loader, num_epochs=20, learning_rate=0.001):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    print('Training...')
    for epoch in range(num_epochs):
        # Training
        model.train()
        train_losses = []
        for inputs, targets in train_loader:
            inputs = inputs.to(device)
            targets = targets.to(device)

            outputs = model(inputs)
            loss = criterion(outputs, targets)
            train_losses.append(loss.item())

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        # Validation
        model.eval()
        val_losses = []
        with torch.no_grad():
            for inputs, targets in val_loader:
                inputs = inputs.to(device)
                targets = targets.to(device)

                outputs = model(inputs)
                loss = criterion(outputs, targets)
                val_losses.append(loss.item())

        print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {np.mean(train_losses)}, Validation Loss: {np.mean(val_losses)}')

    print('Training complete.')


dataset = TopologyOptimizationDataset('top_dataset.h5')

# Split the dataset into train, validation, and test sets
train_size = int(0.7 * len(dataset))
val_size = int(0.15 * len(dataset))
test_size = len(dataset) - train_size - val_size
train_dataset, val_dataset, test_dataset = random_split(dataset, [train_size, val_size, test_size])

# Create data loaders for the datasets
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=True)

# Create the model
model = UNet()

# Train the model
train_model(model, train_loader, val_loader)

# Save the learned weights
torch.save(model.state_dict(), 'model_weights_mymodel.pth')

# Example usage for prediction
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.eval()

accuracies = []
precisions = []
recalls = []
f1s = []
aurocs = []

with torch.no_grad():
    for inputs, targets in test_loader:
        inputs = inputs.to(device)
        targets = targets.to(device)

        prediction = model(inputs)
        binary_prediction = (prediction > 0.5).float()

        accuracy = (binary_prediction == targets).sum().item() / targets.numel()
        accuracies.append(accuracy)

        targets_cpu = targets.cpu().numpy()
        binary_prediction_cpu = binary_prediction.cpu().numpy()

        precision, recall, f1, _ = precision_recall_fscore_support(targets_cpu.ravel(), binary_prediction_cpu.ravel(), average='binary')
        precisions.append(precision)
        recalls.append(recall)
        f1s.append(f1)

        auroc = roc_auc_score(targets_cpu.ravel(), prediction.cpu().numpy().ravel())
        aurocs.append(auroc)

    print('Test Accuracy:', np.mean(accuracies))
    print('Test Precision:', np.mean(precisions))
    print('Test Recall:', np.mean(recalls))
    print('Test F1-score:', np.mean(f1s))
    print('Test AUROC:', np.mean(aurocs))
    
# Visualize a random test input, prediction, and ground truth
with torch.no_grad():
    random_sample_index = random.randint(0, test_size - 1)
    test_input, target = test_dataset[random_sample_index]
    print(f"shape of test_input: {test_input.shape}")
    test_input = test_input.to(device)
    prediction = model(test_input.unsqueeze(0))
    binary_prediction = (prediction > 0.5).float()

random_iteration_index = random.randint(0, 99)  # Assuming each sample has 100 iterations
test_input = test_input[random_iteration_index, :, :].squeeze().cpu().numpy()
binary_prediction = binary_prediction.squeeze().cpu().numpy()
ground_truth = target.squeeze().cpu().numpy()

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(test_input, cmap='gray')
ax[0].set_title(f'Test Input: Sample {random_sample_index}, Iteration {random_iteration_index}')
ax[1].imshow(1-binary_prediction, cmap='binary')
ax[1].set_title('Binary Prediction')
ax[2].imshow(ground_truth, cmap='gray')
ax[2].set_title('Ground Truth')
plt.tight_layout()
plt.show()
````

### Third Neural Network: Another neural network which splits the dataset into test, train and validation set. It will print the accuracies and the output prediction also.

````python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
import h5py
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, roc_auc_score
import random

class UNet(nn.Module):
    def __init__(self):
        super(UNet, self).__init__()
        self.down1 = nn.Sequential(
            nn.Conv2d(100, 32, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, stride=2)
        )
        self.down2 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(128, 128, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, stride=2)
        )
        self.down3 = nn.Sequential(
            nn.Conv2d(128, 256, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(256, 256, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, stride=2)
        )
        self.up3_1 = nn.Sequential(
            nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2),
            nn.ReLU(),
        )
        self.up3_2 = nn.Sequential(
            nn.Conv2d(256, 128, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(128, 64, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
        )
        self.up2_1 = nn.Sequential(
            nn.ConvTranspose2d(64, 32, kernel_size=2, stride=2),
            nn.ReLU(),
        )
        self.up2_2 = nn.Sequential(
            nn.Conv2d(128, 64, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(64, 32, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
        )
        self.up1_1 = nn.Sequential(
            nn.ConvTranspose2d(32, 16, kernel_size=2, stride=2),
            nn.ReLU(),
        )
        self.up1_2 = nn.Sequential(
            nn.Conv2d(132, 32, kernel_size=3, stride=1, padding=1),  # Adjusted from 116 to 132
            nn.ReLU(),
            nn.Conv2d(32, 32, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(32, 1, kernel_size=1, stride=1, padding=0),
            nn.Sigmoid(),
        )

    def forward(self, x):
        x1 = self.down1(x)
        x2 = self.down2(x1)
        x3 = self.down3(x2)

        x3_up = self.up3_1(x3)
        x3_up = F.interpolate(x3_up, size=x2.shape[2:])

        x2_cat = torch.cat([x2, x3_up], dim=1)
        x2_up = self.up3_2(x2_cat)
        x2_up = F.interpolate(x2_up, size=x1.shape[2:])

        x1_cat = torch.cat([x1, x2_up], dim=1)
        x1_up = self.up2_2(x1_cat)
        x1_up = F.interpolate(x1_up, size=x.shape[2:])

        out = self.up1_2(torch.cat([x, x1_up], dim=1))
        return out



class TopologyOptimizationDataset(Dataset):
    def __init__(self, dataset_path):
        with h5py.File(dataset_path, 'r') as h5f:
            self.iters = np.transpose(h5f['iters'], (0, 3, 1, 2))
            self.targets = np.transpose(h5f['targets'], (0, 3, 1, 2))

    def __len__(self):
        return len(self.iters)

    def __getitem__(self, index):
        iter = self.iters[index]
        target = self.targets[index]
        return torch.from_numpy(iter).float(), torch.from_numpy(target).float()


def train_model(model, train_loader, val_loader, num_epochs=20, learning_rate=0.001):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    print('Training...')
    for epoch in range(num_epochs):
        model.train()
        train_losses = []
        for inputs, targets in train_loader:
            inputs = inputs.to(device)
            targets = targets.to(device)

            outputs = model(inputs)
            loss = criterion(outputs, targets)
            train_losses.append(loss.item())

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        model.eval()
        val_losses = []
        with torch.no_grad():
            for inputs, targets in val_loader:
                inputs = inputs.to(device)
                targets = targets.to(device)

                outputs = model(inputs)
                loss = criterion(outputs, targets)
                val_losses.append(loss.item())

        print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {np.mean(train_losses)}, Validation Loss: {np.mean(val_losses)}')

    print('Training complete.')


dataset = TopologyOptimizationDataset('top_dataset.h5')

train_size = int(0.7 * len(dataset))
val_size = int(0.15 * len(dataset))
test_size = len(dataset) - train_size - val_size
train_dataset, val_dataset, test_dataset = random_split(dataset, [train_size, val_size, test_size])

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=True)

model = UNet()

train_model(model, train_loader, val_loader)

torch.save(model.state_dict(), 'model_weights_third.pth')

# Example usage for prediction
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.eval()

accuracies = []
precisions = []
recalls = []
f1s = []
aurocs = []

with torch.no_grad():
    for inputs, targets in test_loader:
        inputs = inputs.to(device)
        targets = targets.to(device)

        prediction = model(inputs)
        binary_prediction = (prediction > 0.5).float()

        accuracy = (binary_prediction == targets).sum().item() / targets.numel()
        accuracies.append(accuracy)

        targets_cpu = targets.cpu().numpy()
        binary_prediction_cpu = binary_prediction.cpu().numpy()

        precision, recall, f1, _ = precision_recall_fscore_support(targets_cpu.ravel(), binary_prediction_cpu.ravel(), average='binary')
        precisions.append(precision)
        recalls.append(recall)
        f1s.append(f1)

        auroc = roc_auc_score(targets_cpu.ravel(), prediction.cpu().numpy().ravel())
        aurocs.append(auroc)

    print('Test Accuracy:', np.mean(accuracies))
    print('Test Precision:', np.mean(precisions))
    print('Test Recall:', np.mean(recalls))
    print('Test F1-score:', np.mean(f1s))
    print('Test AUROC:', np.mean(aurocs))
    
# Visualize a random test input, prediction, and ground truth
with torch.no_grad():
    random_sample_index = random.randint(0, test_size - 1)
    test_input, target = test_dataset[random_sample_index]
    print(f"shape of test_input: {test_input.shape}")
    test_input = test_input.to(device)
    prediction = model(test_input.unsqueeze(0))
    binary_prediction = (prediction > 0.5).float()

random_iteration_index = random.randint(0, 99)  # Assuming each sample has 100 iterations
test_input = test_input[random_iteration_index, :, :].squeeze().cpu().numpy()
binary_prediction = binary_prediction.squeeze().cpu().numpy()
ground_truth = target.squeeze().cpu().numpy()

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(test_input, cmap='gray')
ax[0].set_title(f'Test Input: Sample {random_sample_index}, Iteration {random_iteration_index}')
ax[1].imshow(1-binary_prediction, cmap='binary')
ax[1].set_title('Binary Prediction')
ax[2].imshow(ground_truth, cmap='gray')
ax[2].set_title('Ground Truth')
plt.tight_layout()
plt.show()
````

# Transfer Learning

````python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.models as models
from torch.utils.data import Dataset, DataLoader
import h5py
import numpy as np
import random

# Load the dataset from top_dataset.h5
class TopologyOptimizationDataset(Dataset):
    def __init__(self, dataset_path):
        with h5py.File(dataset_path, 'r') as h5f:
            self.iters = np.transpose(h5f['iters'], (0, 3, 1, 2))
            self.targets = np.transpose(h5f['targets'], (0, 3, 1, 2))

    def __len__(self):
        return len(self.iters)

    def __getitem__(self, index):
        iter = self.iters[index]
        target = self.targets[index]
        return torch.from_numpy(iter).float(), torch.from_numpy(target).float()

# Create the modified VGG16 model for transfer learning
def create_vgg16_model():
    vgg16 = models.vgg16(pretrained=True)
    
    # Replace the first layer to handle single-channel input
    vgg16.features[0] = nn.Conv2d(1, 64, kernel_size=3, padding=1)
    
    # Modify the last layer for regression with 1 channel output
    in_features = vgg16.classifier[-1].in_features
    vgg16.classifier[-1] = nn.Linear(in_features, 40 * 40)

    return vgg16

# Train the model with transfer learning
def train_model(model, train_loader, num_epochs=20, learning_rate=0.001):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    print('Training...')
    for epoch in range(num_epochs):
        model.train()
        train_losses = []
        for inputs, targets in train_loader:
            inputs = inputs.to(device)
            targets = targets.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()

            train_losses.append(loss.item())

        print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {np.mean(train_losses)}')

    print('Training complete.')

if __name__ == '__main__':
    dataset = TopologyOptimizationDataset('top_dataset.h5')

    # Define the DataLoader with batch_size
    batch_size = 64
    train_loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

    # Create the modified VGG16 model
    model = create_vgg16_model()

    # Debugging print statements to check the model architecture
    print(model)
    print(model.features[0])  # Check the first layer of the model

    # Train the model with transfer learning
    train_model(model, train_loader)

    # Save the trained model
    torch.save(model.state_dict(), 'transfer_learning_model.pth')
````
